```python
t = linspace(0, 2*pi, 50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```

```python
subplot(1, 2, 1)
plot(x)
subplot(1, 2, 2)
plot(y)
```

```python
plot(x)
plot(y)
```

```python
plot(x)
hold(False)
plot(y)
hold(True)
```

```python
plot(x, label='sin')
plot(y, label='cos')
legend()
```

```python
plot(x)
plot(y)
legend(['sin', 'cos'])
```

```python
plot(x, sin(x))
xlabel('radians')
ylabel('amplitude', fontsize='large')
title('Sin(x)')
```
